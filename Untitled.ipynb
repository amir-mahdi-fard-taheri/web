```python
!pip install numpy opencv-python matplotlib
import cv2
import numpy as np
import matplotlib.pyplot as plt
import zlib
from google.colab import files
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import mean_squared_error as mse
from skimage.metrics import structural_similarity as ssim

### مرحله ۲: دریافت تصویر از کاربر و تبدیل به سطح خاکستری
print("لطفاً تصویر خود را آپلود کنید:")
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# بارگذاری تصویر و تبدیل به سطح خاکستری
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
plt.title('User Uploaded Image')
plt.axis('off')
plt.show()

### مرحله ۳: فشرده‌سازی تصویر با LZMC
def compress_image(image):
    flattened_data = image.flatten()
    compressed_data = zlib.compress(flattened_data)
    return compressed_data

compressed_image = compress_image(image)
print(f"Compressed size: {len(compressed_image)} bytes")

### مرحله ۴: رمزگذاری تصویر با XOR و نقشه‌های آشوبی
def chaotic_encryption(image, key=42):
    encrypted_image = np.bitwise_xor(image, key)
    return encrypted_image

encrypted_image = chaotic_encryption(image)

plt.imshow(encrypted_image, cmap='gray')
plt.title('Encrypted Image')
plt.axis('off')
plt.show()

### مرحله ۵: ارزیابی کیفیت پردازش تصویر
def evaluate_quality(original, processed):
    psnr_value = psnr(original, processed)
    mse_value = mse(original, processed)
    ssim_value = ssim(original, processed)
    return psnr_value, mse_value, ssim_value

psnr_value, mse_value, ssim_value = evaluate_quality(image, encrypted_image)
print(f"PSNR: {psnr_value:.2f} dB")
print(f"MSE: {mse_value:.4f}")
print(f"SSIM: {ssim_value:.4f}")

### مرحله ۶: بازیابی تصویر رمزگذاری‌شده
def chaotic_decryption(encrypted_image, key=42):
    decrypted_image = np.bitwise_xor(encrypted_image, key)
    return decrypted_image

decrypted_image = chaotic_decryption(encrypted_image)

plt.imshow(decrypted_image, cmap='gray')
plt.title('Decrypted Image')
plt.axis('off')
plt.show()

### مرحله ۷: نمایش تفاوت بین تصاویر پردازش‌شده
fig, axs = plt.subplots(1, 3, figsize=(15,5))

axs[0].imshow(image, cmap='gray')
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(encrypted_image, cmap='gray')
axs[1].set_title('Encrypted Image')
axs[1].axis('off')

axs[2].imshow(decrypted_image, cmap='gray')
axs[2].set_title('Decrypted Image')
axs[2].axis('off')

plt.show()
```
